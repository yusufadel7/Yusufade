# Corrolation between features

```python
# do heatmap for numeric columns (, ,)
import seaborn as sns
import matplotlib.pyplot as plt

# Select specific numeric columns
columns_of_interest = ['Credit_Score','Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_of_Loan', 'Delay_from_due_date', 'Outstanding_Debt', 'Amount_invested_monthly', 'Monthly_Balance']
numeric_df = df[columns_of_interest]

# Ensure these columns are numeric and convert if necessary
numeric_df = numeric_df.apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN values to ensure we can compute the correlation matrix
numeric_df = numeric_df.dropna()

# Calculate correlation
heatmap = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap for numerical columns')
plt.show()
```

## there appear to be strong correlations between several features in the provided correlation matrix:

-Monthly Inhand Salary has a strong positive correlation with Monthly_Balance (0.71) and Amount_invested_monthly (0.64), suggesting that individuals with higher salaries tend to have higher balances and invest more.

-Outstanding_Debt has strong negative correlations with Credit_Score (-0.55), Delay_from_due_date (-0.53), and Monthly_Balance (-0.36), indicating that individuals with higher debt tend to have lower credit scores, are more likely to delay payments, and have lower monthly balances.

-Delay_from_due_date has a strong positive correlation with Outstanding_Debt (0.58), suggesting that individuals who delay payments are more likely to have higher outstanding debt.

## Conclusion:

#### This dataset encompasses various attributes that collectively provide a comprehensive view of an individual's financial profile and credit behavior.




-Personal Information:
Month: Indicates the time frame of the data, which can help identify trends over time.
Name, Age, SSN, Occupation: These columns provide demographic details that are crucial for understanding the socio-economic context of the individuals.



-Income and Salary:
Annual_Income & Monthly_Inhand_Salary: These metrics offer insights into the individual’s earning capacity, which is essential for assessing financial stability and budgeting.




-Debt and Loan Information:
Num_of_Loan, Type_of_Loan, Outstanding_Debt: These attributes reflect the individual's borrowing behavior and current debt obligations, which are critical for evaluating financial health.




-Delay_from_due_date:
This column indicates payment punctuality, directly influencing creditworthiness.




-Credit Behavior:
Changed_Credit_Limit, Num_Credit_Inquiries, Credit_Utilization_Ratio, Credit_History_Age: These metrics provide insights into the individual’s credit management practices, including how they utilize credit, their history of inquiries, and the length of their credit history.




-Payment_of_Min_Amount: This indicates whether the individual meets minimum payment requirements, which is vital for maintaining a good credit score.




-Financial Management:
Total_EMI_per_month, Amount_invested_monthly: These columns help analyze the monthly financial commitments and investments, providing a picture of cash flow management.




-Monthly_Balance: This reflects the individual’s financial status after accounting for income and expenses.




-Credit Score and Payment Behavior:
Credit_Score, Payment_Behaviour: These are critical indicators of creditworthiness and payment reliability, which lenders use to assess risk.
